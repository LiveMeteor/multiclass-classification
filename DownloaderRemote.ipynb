#### Download the dataset from the remote server and save them to local folder

```python
import pandas as pd

# Kaggle: https://www.kaggle.com/datasets/benjaminkz/places365/data
splits = {'train': 'data/train-00000-of-00001.parquet', 'test': 'data/test-00000-of-00001.parquet'}

df_train = pd.read_parquet("hf://datasets/dpdl-benchmark/places365-mini-sample-hard/" + splits["train"])
df_train.to_parquet("./data/places365-mini-sample-hard-train.parquet")

df_test = pd.read_parquet("hf://datasets/dpdl-benchmark/places365-mini-sample-hard/" + splits["test"])
df_test.to_parquet("./data/places365-mini-sample-hard-test.parquet")

print(f'Preview of training data:\n{df_train.head()}')
print(f'Shape of training and test data: {df_train.shape}, {df_test.shape}')

print(f'Label distribution in training data:\n{df_train["label"].value_counts()}')
print(f'Label distribution in test data:\n{df_test["label"].value_counts()}')
```

#### Loading the images from the local files and save them in different directories respectively.

```python
import os
import io
from PIL import Image
import pandas as pd

class_names = [
    'Boxing_Stage',       # 0
    'Inside_a_Car',       # 1
    'Airplane_Cockpit',   # 2
    'Forest',             # 3
    'Outside_a_Car',      # 4
    'Conference_Room',    # 5
    'Rehearsal_Room',     # 6
    'Stage',              # 7
    'Performance',        # 8
    'Conference_Hall',    # 9
]

def save_images_by_label(df, split_dir):
    """Save images from a DataFrame into subdirectories named by label."""
    counters = {i: 0 for i in range(len(class_names))}
    for _, row in df.iterrows():
        label = row["label"]
        image_bytes = row["image"]["bytes"]
        class_dir = os.path.join(split_dir, f"{label}_{class_names[label]}")
        os.makedirs(class_dir, exist_ok=True)
        img = Image.open(io.BytesIO(image_bytes))
        img_path = os.path.join(class_dir, f"{counters[label]:05d}.jpg")
        img.save(img_path)
        counters[label] += 1
    return counters

df_train = pd.read_parquet("./data/places365-mini-sample-hard-train.parquet")
df_test = pd.read_parquet("./data/places365-mini-sample-hard-test.parquet")

train_counts = save_images_by_label(df_train, "./data/images/train")
test_counts  = save_images_by_label(df_test,  "./data/images/test")

print("Training images saved per class:")
for label, count in train_counts.items():
    print(f"  {label} - {class_names[label]}: {count}")

print("\nTest images saved per class:")
for label, count in test_counts.items():
    print(f"  {label} - {class_names[label]}: {count}")
```
